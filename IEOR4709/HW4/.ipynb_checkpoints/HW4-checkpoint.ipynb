```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

import statsmodels.api as sm
```

```python
# Source taken from Google Finance
df = pd.read_csv('StockPrices.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].apply(lambda x: x.month)

month_end_df = df[df['Month End'] == 1].copy()
month_end_df['SPX_Rets'] = month_end_df['SPX'].pct_change()
month_end_df['IBM_Rets'] = month_end_df['IBM'].pct_change()

month_end_df['const'] = 1
```

```python
month_end_df.head()
```

```python
fig, ax = plt.subplots(1, 2)

fig.set_figheight(6)
fig.set_figwidth(15)

on_trigger = month_end_df[month_end_df['Month'] == 12].dropna()
off_trigger = month_end_df[month_end_df['Month'] != 12].dropna()

# Observing only December Returns
ax[0].spines[['right', 'top']].set_visible(False)
ax[0].scatter(on_trigger['SPX_Rets'], on_trigger['IBM_Rets'], color='r', alpha=0.5, label='Dec Returns')
ax[0].set_xlabel('SPX Daily Returns')
ax[0].set_ylabel('IBM Daily Returns')

model = sm.OLS(on_trigger['IBM_Rets'], on_trigger[['const', 'SPX_Rets']])
results = model.fit()
X = np.linspace(start=on_trigger['SPX_Rets'].min(), stop=on_trigger['SPX_Rets'].max())
ax[0].plot(X, results.params.const + X*results.params.SPX_Rets, color='r')
ax[0].annotate(f'alpha={results.params.const:.2f}\nbeta={results.params.SPX_Rets:.2f}', 
               xy=(on_trigger['SPX_Rets'].max()*0.6, on_trigger['IBM_Rets'].min()*0.9),
               weight="bold", fontsize=12)

# Observing only January to November Returns
ax[1].spines[['right', 'top']].set_visible(False)
ax[1].scatter(off_trigger['SPX_Rets'], off_trigger['IBM_Rets'], color='b', alpha=0.5, label='Jan-Nov Returns')
ax[1].set_xlabel('SPX Daily Returns')
ax[1].set_ylabel('IBM Daily Returns')

model = sm.OLS(off_trigger['IBM_Rets'], off_trigger[['const', 'SPX_Rets']])
results = model.fit()
X = np.linspace(start=off_trigger['SPX_Rets'].min(), stop=off_trigger['SPX_Rets'].max())
ax[1].plot(X, results.params.const + X*results.params.SPX_Rets, color='b')
ax[1].annotate(f'alpha={results.params.const:.2f}\nbeta={results.params.SPX_Rets:.2f}', 
               xy=(off_trigger['SPX_Rets'].max()*0.6, off_trigger['IBM_Rets'].min()*0.9),
               weight="bold", fontsize=12)

ax[0].legend()
ax[1].legend()
plt.show()
```
