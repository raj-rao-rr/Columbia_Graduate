```python
import pandas as pd
import numpy as np 
import gurobipy as gp

import seaborn as sns
import matplotlib.pyplot as plt
```

**We begin by loading in our data values for both refugees and the transportation network**

```python
refugee = pd.read_excel('rao_rajesh_assign_hw3.xlsx', sheet_name=1)
network = pd.read_excel('rao_rajesh_assign_hw3.xlsx', sheet_name=0)
```

```python
refugee.head()
```

## Question 1

```python
# instantiate the variable assignment
routes = (network['From'] + '_' + network['To']).values

# path dependent constraints
person_limit = dict(zip(routes, network['Person Limit']))
vehicle_nums = dict(zip(routes, network['Vehicle Number']))

# node definitions
nodes = np.concatenate((network.From.unique(), network.To.unique())) 
starting_point = 'Gaza'
ending_point = 'Sofia'
```

```python
node = 'Cairo'

display(network[network.To == node])
display(network[network.From == node])
```

```python
def path_creator(df:pd.DataFrame)-> list:
    """Takes a dataframe and creates a path connector"""
    return (df['From'] + '_' + df['To']).values
```

```python
m1 = gp.Model('question1')

paths = {i: m1.addVar(vtype=gp.GRB.INTEGER, name=f'{i}') for i in routes}

# objective of reaching the maximum flow within a network
m1.setObjective(sum([paths[i] for i in path_creator(network[network.To == ending_point])]), gp.GRB.MAXIMIZE)

# flow constraints based on network direction
for node in nodes[~np.isin(nodes, [starting_point, ending_point])]:
    from_path = network[network.From == node]
    to_path = network[network.To == node]
    
    X = sum(paths[i] for i in path_creator(from_path))
    Y = sum(paths[i] for i in path_creator(to_path))
    m1.addConstr(X == Y, name=f'{node}_flow_constr')

# flow limitation by transport
for p in paths: 
    m1.addConstr(paths[p] <= vehicle_nums[p] * person_limit[p], name=f'{p}_person_vehicle_constr')
    
m1.optimize()

if m1.status == gp.GRB.OPTIMAL:
    print(f'\nOf the 2000 Gazan refugees, only {m1.ObjVal} will be able to reach Sofia within a day\nsee flow path below:')
    
    flows = [paths[p].x for p in paths]
    out_df = pd.DataFrame({
        'Path': paths.keys(), 
        'People Flow': flows, 
        'Person Limit': person_limit.values(), 
        'Max # of Vehicles': vehicle_nums.values(),
        'Vehicles Used': np.ceil(np.array(flows) / np.array(list(person_limit.values()))),
    })
    
    display(out_df)
    
```

## Question 2

```python
fix, axe = plt.subplots()
sns.kdeplot(data=refugee, x="Age", hue="Gender", ax=axe)
axe.axvline(refugee[refugee.Gender == 'Male'].Age.mean(), ls='--', color='tab:blue')
axe.axvline(refugee[refugee.Gender == 'Female'].Age.mean(), ls='--', color='tab:orange')
plt.show()
```

```python
refugee_age = dict(zip(range(refugee.shape[0]), refugee['Age']))
```

```python
m2 = gp.Model('question2')

assign = {i: m2.addVar(vtype=gp.GRB.BINARY, name=f'{refugee.iloc[i].Name}') for i in range(refugee.shape[0])}

# objective of reaching the maximum flow within a network
m2.setObjective(sum(assign[i] * refugee_age[i] for i in range(refugee.shape[0])), gp.GRB.MINIMIZE)

m2.addConstr(sum(assign[i] for i in range(refugee.shape[0])) == 400)

m2.optimize()

if m2.status == gp.GRB.OPTIMAL:
    print(f'\nThere are {sum(assign[i].x for i in range(refugee.shape[0]))} people evacuated')
    
    ppl_assignment = [assign[i].x for i in range(refugee.shape[0])]
    out_df = pd.DataFrame({
        'People': [refugee.iloc[i].Name for i in range(refugee.shape[0])],
        'Assignment': ppl_assignment,
        'Age': refugee.Age
    })
    
    display(out_df[out_df.Assignment == 1])
    
    x = out_df[out_df.Assignment == 1]['Age'].mean()
    print(f'The average age of the evacuees is {x}')
```

## Question 3

```python
# construct a flag to determine whethre a refugee is a child or not
refugee['Child_Flag'] = refugee['Age'].apply(lambda x: 1 if x < 18 else 0)
refugee['Adult_Flag'] = refugee['Child_Flag'].apply(lambda x: 0 if x == 1 else 1)
```

```python
refugee.head()
```

```python
m3 = gp.Model('question3')

assign = {i: m3.addVar(vtype=gp.GRB.BINARY, name=f'{refugee.iloc[i].Name}') for i in range(refugee.shape[0])}

# objective of reaching the maximum flow within a network
m3.setObjective(sum(assign[i] * refugee_age[i] for i in range(refugee.shape[0])), gp.GRB.MINIMIZE)

m3.addConstr(sum(assign[i] for i in range(refugee.shape[0])) == 400)
m3.addConstr(sum(assign[i]*refugee.iloc[i].Child_Flag for i in range(refugee.shape[0])) == \
             4*sum(assign[i]*refugee.iloc[i].Adult_Flag for i in range(refugee.shape[0])))

m3.optimize()

if m3.status == gp.GRB.OPTIMAL:
    print(f'\nThere are {sum(assign[i].x for i in range(refugee.shape[0]))} people evacuated')
    
    ppl_assignment = [assign[i].x for i in range(refugee.shape[0])]
    out_df = pd.DataFrame({
        'People': [refugee.iloc[i].Name for i in range(refugee.shape[0])],
        'Assignment': ppl_assignment,
        'Age': refugee.Age,
        'Child_Flag': refugee.Child_Flag
    })
    
    display(out_df[out_df.Assignment == 1])
    
    x = out_df[out_df.Assignment == 1]['Age'].mean()
    adult_f = out_df[(out_df.Assignment == 1) & (out_df.Child_Flag == 0)]['Age']
    child_f = out_df[(out_df.Assignment == 1) & (out_df.Child_Flag == 1)]['Age']
    
    print(f'Number of adults {adult_f.count()}, Number of children {child_f.count()}')
    print(f'The average age of the evacuees is {x} (adults:{adult_f.mean()}, children:{child_f.mean()})')
```

## Question 4

```python
def age_scalar(x:int) -> float:
    
    if 0 <= x <= 3:
        return 0.25
    elif 4 <= x <= 14:
        return 0.5
    else:
        return 1
```

```python
# scalar adjustment based on the refugee age
refugee['Age_Scalar'] = refugee['Age'].apply(lambda x: age_scalar(x))
```

```python
m4 = gp.Model('question4')

assign = {i: m4.addVar(vtype=gp.GRB.BINARY, name=f'{refugee.iloc[i].Name}') for i in range(refugee.shape[0])}

# objective of reaching the maximum flow within a network
m4.setObjective(sum(assign[i] * refugee_age[i] for i in range(refugee.shape[0])), gp.GRB.MINIMIZE)

m4.addConstr(sum(assign[i]*refugee.iloc[i].Age_Scalar for i in range(refugee.shape[0])) == 400)
m4.addConstr(sum(assign[i]*refugee.iloc[i].Child_Flag for i in range(refugee.shape[0])) == \
             4*sum(assign[i]*refugee.iloc[i].Adult_Flag for i in range(refugee.shape[0])))
m4.optimize()

if m4.status == gp.GRB.OPTIMAL:
    print(f'\nThere are {sum(assign[i].x for i in range(refugee.shape[0]))} people evacuated')
    
    ppl_assignment = [assign[i].x for i in range(refugee.shape[0])]
    out_df = pd.DataFrame({
        'People': [refugee.iloc[i].Name for i in range(refugee.shape[0])],
        'Assignment': ppl_assignment,
        'Age': refugee.Age,
        'Child_Flag': refugee.Child_Flag
    })
    
    display(out_df[out_df.Assignment == 1])
    
    x = out_df[out_df.Assignment == 1]['Age'].mean()
    adult_f = out_df[(out_df.Assignment == 1) & (out_df.Child_Flag == 0)]['Age']
    child_f = out_df[(out_df.Assignment == 1) & (out_df.Child_Flag == 1)]['Age']
    
    print(f'Number of adults {adult_f.count()}, Number of children {child_f.count()}')
    print(f'The average age of the evacuees is {round(x,2)} (adults:{round(adult_f.mean(), 2)}, children:{round(child_f.mean(), 2)})')
```
