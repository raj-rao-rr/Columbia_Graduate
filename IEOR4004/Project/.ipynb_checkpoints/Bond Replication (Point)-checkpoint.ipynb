# Portfolio Replication Problem

A bond benchmark portfolio with **N** securities will have an aggregate yield and duration modified duration measurement. 

We would like to create a subset (“*replication*”) portfolio with a smaller numbre of bonds from the broader benchmark, such that we minimize the yield and duration differential between the provided benchmark portfolio and the replication portfolio. We will be using the J.P. Morgan Government Bond–Emerging Market Index (GBI-EM) series as our bond benchmark for this excercise in portfolio replication.

However, we will be dividing this problem into two parts **A)** where we select the most liquid subset of tradable bonds from the benchmark to be in the replication portfolio, and **B)** where we will determine the appropriate weight for each bond such that we match the benchmark's yield and duration. 

```python
import math
import pandas as pd 
import numpy as np 
# import gurobipy as gp 
import matplotlib.pyplot as plt
from scipy.optimize import minimize, linprog, LinearConstraint
```

```python
# Load in the J.P. Morgan GBI-EM Global Bond Index composition
df = pd.read_csv('JPM_COMP_GBI_EMGLOBAL_DIV_2023-07-31.csv')
```

**Liquidity Scores** corresponds with the tradability of a given bond within a country of issuance, e.g. CND10003VNX4 China bond has a liquidity score of 1.0, implying that it is more liquid compared to other bonds from China in the benchmark. There are three liquidity score assignments 1, 0.5, and 0.33. 	  

```python
# Define date-oriented figures 
df['Issue_Dt'] = pd.to_datetime(df['Issue_Dt'])
df['Maturity'] = pd.to_datetime(df['Maturity'])
df['Days_Since_Issuance'] = (pd.Timestamp.today() - df['Issue_Dt']).apply(lambda x: x.days)

# Detemine the Liqudity score for a given bond
cntry_agg_wgt = (df.groupby('Country')['Weight % - Daily USD Rtrn'].sum().sort_values(ascending=False)/100).to_dict()
cntry_agg_cnt = df.groupby('Country')['Weight % - Daily USD Rtrn'].count().to_dict()

# Compute the weight of a bond within country issuer
df['Country_Weight'] = df['Country'].apply(lambda x: cntry_agg_wgt[x])
df['Intra_Country_Weight'] = df['Weight % - Daily USD Rtrn'] / df['Country_Weight'] 

df['Liquidity_Score'] = df['Intra_Country_Weight'] / df['Days_Since_Issuance']

# List of countries
countries = df.Country.unique()
```

```python
df.head()
```

```python
n = 50
```

```python
def cnty_flr_assign(cntry_wgt:dict, n:int=50):
    """
    Country Floor Assignment
    """
    
    min_allocation = lambda x: np.floor(cntry_wgt[x]*n)
    flr_def = dict(zip(cntry_wgt.keys(), map(lambda x: 1 if min_allocation(x) == 0 else min_allocation(x), cntry_wgt)))
    
    if sum(flr_def.values()) < 50:
        for key, val in flr_def.items():
            flr_def[key] += 1
            
            if sum(flr_def.values()) == 50:
                return flr_def 
    
    elif sum(flr_def.values()) > 50:
        for key, val in flr_def.items():
            flr_def[key] -= 1
            
            if sum(flr_def.values()) == 50:
                return flr_def 
    
    else:
        return flr_def
```

## A) Subset Determination
**Linear Assignment Problem: Determing the most liquid subset of bonds to include in the replication basket**

```python
assign_binary = cnty_flr_assign(cntry_agg_wgt)
```

```python
assign_binary
```

```python
def bond_subset(df, selection, dimension):
    
    concat_slice = [0]*dimension.size
    
    for idx, c in enumerate(dimension):
        cntry_filter = df[df['Country'] == c]
        cntry_sorted = cntry_filter.sort_values(by='Liquidity_Score')
        concat_slice[idx] = cntry_sorted.iloc[-int(selection[c]):]

    return pd.concat(concat_slice).reset_index(drop=True)
```

```python
replication_portfolio = bond_subset(df, assign_binary, countries)
```

## B) Replication Weights

**Linear Weighting Problem: Determining the appropriate weight to assign each bond such that we mirror the benchmark**

```python
bench_yield = df['Yield'].values
bench_duration = df['Mac Dur'].values
bench_weight = df['Weight % - Daily Stats'].values
```

```python
yld_value = sum(bench_weight * bench_duration * bench_yield) / sum(bench_weight * bench_duration)
dur_value = sum(bench_weight/100 * bench_duration)
print(yld_value, dur_value)
```

```python
cntry_bond_map = {c: replication_portfolio[replication_portfolio.Country == c].index.values for c in countries}
```

```python
# Consider a minimization problem with several constraints
def opt_func(x, y, d, bench_yield, bench_duration):
    yld = yield_calc(x, y, d)
    dur = duration_calc(x, d)
    return (yld - bench_yield)**2 + (dur - bench_duration)**2

def yield_calc(x, y, d):
    return sum(x * y * d) / sum(x * d)
    
def duration_calc(x, d):
    return sum(x * d)

def single_constraint(cntry_ref, selection):
    def inner(x):
        return 1e-6 - abs(cntry_ref - x[selection].sum())
    return inner

cons = []
total_v = sum(list(cntry_agg_wgt.values()))

print(f'CONSTRAINTS')
# # CONSTRAINT (1): Full Investment constraint 
cons.append({'type': 'eq', 
             'fun': lambda x: 1 - x.sum()})


# CONSTRAINT (2): Diversification constraint, no country weight can exceed its benchmark weight
for c in countries:
    print(f'\t{c} => index over {cntry_bond_map[c]} sums to => {cntry_agg_wgt[c]}')
    cons.append({'type': 'ineq', 
                 'fun': single_constraint(cntry_agg_wgt[c], cntry_bond_map[c])})


# CONSTRAINT (3): No shorting constraint, 0 <= x <= 1
bnds = tuple((0, 1) for i in range(replication_portfolio.shape[0]))


print(f'\nOPTIMIZE')
x0 = np.zeros(shape=(50, )) + 1/n

res = minimize(fun=opt_func, 
               x0=x0, 
               method='trust-constr', 
               bounds=bnds, 
               constraints=tuple(cons), 
               args=(replication_portfolio['Yield'].values, 
                     replication_portfolio['Mac Dur'].values,
                     yld_value, 
                     dur_value),
               tol=1e-4) 


for c in countries:
    print(f'\t{c} =>\tReplication Weight {cntry_agg_wgt[c] - res.x[cntry_bond_map[c]].sum()}')

print(f'\nTotal Portfolio Weight {res.x.sum()}')
y = replication_portfolio['Yield'].values
d = replication_portfolio['Mac Dur'].values
print(f'\tReplication Yld: {yield_calc(res.x, y, d)}, Bench Yld: {yld_value}')
print(f'\tReplication Dur: {duration_calc(res.x, d)}, Bench Dur: {dur_value}')
```
