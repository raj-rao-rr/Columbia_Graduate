```python
import pandas as pd 
import numpy as np 
import gurobipy as gp 
```

### Questoin 2

```python
x = [1, 3, 5, 7, 2, 6]
y = [2, 1, 3, 2, 2, 5]

distance = [[( (x[i]-x[j])**2+ (y[i]-y[j])**2 )**0.5 for j in range(len(x))] for i in range(len(x))]
n = len(distance)
```

```python
np.array(distance)
```

```python
# no-SUBTOURS
m = gp.Model('tsp')

# define the binary decision variables
alloc = {f'{i},{j}': m.addVar(vtype=gp.GRB.BINARY, name=f'({i}, {j})') for i in range(n) for j in range(n)}

m.setObjective(sum(alloc[p]* distance[int(p.split(',')[0])][int(p.split(',')[1])] for p in alloc), 
               gp.GRB.MINIMIZE)

# flow constraints 
for j in range(n):
    m.addConstr(sum(alloc[f'{i},{j}'] for i in range(n)) == 1)
    
for i in range(n):
    m.addConstr(sum(alloc[f'{i},{j}'] for j in range(n)) == 1)

for i in range(n):
    m.addConstr(alloc[f'{i},{i}'] == 0)

m.optimize()

if m.status == gp.GRB.OPTIMAL:
    
    print('Optimal Sequence (including sub-tours):')
    for p in alloc:
        if alloc[p].x > 0:
            print(f'{p} => {alloc[p].x}')
```

## Question 4

```python
m = gp.Model('tsp')

# define the binary decision variables
alloc = {f'{i},{j}': m.addVar(vtype=gp.GRB.BINARY, name=f'({i}, {j})') for i in range(n) for j in range(n)}

m.setObjective(sum(alloc[p]* distance[int(p.split(',')[0])][int(p.split(',')[1])] for p in alloc), 
               gp.GRB.MINIMIZE)

u = {}
for i in range(n):
    if (i > 0):
        u[i] = m.addVar(vtype=gp.GRB.CONTINUOUS, lb=1.0, ub=n-1, obj=0, name=f'u({i})')
    else:
        u[i] = m.addVar(vtype=gp.GRB.CONTINUOUS, lb=0, ub=0, obj=0, name=f'u({i})')

# flow constraints 
for j in range(n):
    m.addConstr(sum(alloc[f'{i},{j}'] for i in range(n)) == 1)
    
for i in range(n):
    m.addConstr(sum(alloc[f'{i},{j}'] for j in range(n)) == 1)

for i in range(n):
    m.addConstr(alloc[f'{i},{i}'] == 0)

# sub-tour handling
for i in range(n-1):
    for j in range(n-1):
        if (i != j):
            m.addConstr(u[j+1] >= u[i+1] + 1 - (n-1)*(1-alloc[f'{i},{j}']))
    
m.optimize()

if m.status == gp.GRB.OPTIMAL:
    
    print('Optimal Sequence (including sub-tours):')
    u_order_vector = [int(u[i].x) for i in range(n)]
    print('\tU Order vector ', u_order_vector)
    points = list(zip(x, y))
    tsp_points = [points[i] for i in u_order_vector]
    print('\tData Points (nodes) ', tsp_points)
```
