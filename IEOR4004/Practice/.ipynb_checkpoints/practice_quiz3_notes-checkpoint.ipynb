### Question 1

```python
city_map = {(1,1):0, (1,2):10, (1,3):20, (1,4):30, (1,5):30, (1,6):20,
            (2,1):10, (2,2):0, (2,3):25, (2,4):35, (2,5):20, (2,6):10,
            (3,1):20, (3,2):25, (3,3):0, (3,4):15, (3,5):30, (3,6):20,
            (4,1):30, (4,2):35, (4,3):15, (4,4):0, (4,5):15, (4,6):25,
            (5,1):30, (5,2):20, (5,3):30, (5,4):15, (5,5):0, (5,6):14,
            (6,1):20, (6,2):10, (6,3):20, (6,4):25, (6,5):14, (6,6):0}
```

```python
m = gp.Model('question1')

alloc = {i: m.addVar(vtype=gp.GRB.BINARY) for i in range(1, 6+1)}

m.setObjective(sum(alloc[i] for i in alloc), gp.GRB.MINIMIZE)

m.addConstr(alloc[1]+alloc[2] >= 1)
m.addConstr(alloc[1]+alloc[2]+alloc[6] >= 1)
m.addConstr(alloc[3]+alloc[4] >= 1)
m.addConstr(alloc[3]+alloc[4]+alloc[5] >= 1)
m.addConstr(alloc[4]+alloc[5]+alloc[6] >= 1)
m.addConstr(alloc[2]+alloc[5]+alloc[6] >= 1)

m.optimize()

if m.status == gp.GRB.OPTIMAL:
    
    for i in alloc:
        print(f'{i} = {alloc[i].x}')
```

### Question 2

```python
bakery_map = {
    ('Dough', 'Tiny'): 90, ('Dough', 'Small'): 125, ('Dough', 'Midi'): 175, ('Dough', 'Large'): 250, 
    ('Icing', 'Tiny'): 35, ('Icing', 'Small'): 50, ('Icing', 'Midi'): 75, ('Icing', 'Large'): 90,
    ('Labor', 'Tiny'): 1, ('Labor', 'Small'): 1, ('Labor', 'Midi'): 1, ('Labor', 'Large'): 1,
    ('Profit', 'Tiny'): 1, ('Profit', 'Small'): 2, ('Profit', 'Midi'): 3, ('Profit', 'Large'): 4,
}
```

**Leverage the binary in expressoin**

```python
m = gp.Model('question2')

alloc = {i: m.addVar(vtype=gp.GRB.INTEGER) for i in ['Tiny', 'Small', 'Midi', 'Large']}
selection = {i: m.addVar(vtype=gp.GRB.BINARY) for i in ['Tiny', 'Small', 'Midi', 'Large']}

m.setObjective(sum(alloc[i]*bakery_map['Profit', i]*selection[i] for i in alloc), gp.GRB.MAXIMIZE)

# all supply constraints
m.addConstr(sum(alloc[i]*bakery_map['Dough', i] for i in alloc) <= 65000)
m.addConstr(sum(alloc[i]*bakery_map['Icing', i] for i in alloc) <= 25000)
m.addConstr(sum(alloc[i]*bakery_map['Labor', i] for i in alloc) <= 800)

M = 1e7
for i in alloc:
    # pure binary method (selection on/off)
    m.addConstr(alloc[i] >= 600 * selection[i])
    
#     # big M-method method
#     m.addConstr(alloc[i] >= 600 - M * (1-selection[i]))
#     m.addConstr(alloc[i] <= M * selection[i])

m.optimize()

if m.status == gp.GRB.OPTIMAL:
    
    for i in alloc:
        print(f'{i} = {alloc[i].x},{selection[i].x} ')
```

**Big M-method**

```python
from gurobipy import Model , GRB

# Create a new model
m = Model("bakery_model")
# Add variables
x1 = m.addVar(vtype=GRB.INTEGER , name="x1")
x2 = m.addVar(vtype=GRB.INTEGER , name="x2")
x3 = m.addVar(vtype=GRB.INTEGER , name="x3")
x4 = m.addVar(vtype=GRB.INTEGER , name="x4")
y1 = m.addVar(vtype=GRB.BINARY , name="y1")
y2 = m.addVar(vtype=GRB.BINARY , name="y2")
y3 = m.addVar(vtype=GRB.BINARY , name="y3")
y4 = m.addVar(vtype=GRB.BINARY , name="y4")

# Big M
M = 10000 # You might want to adjust this based on the problem context
# Set the objective
m.setObjective(x1*y1 + 2*x2*y2 + 3*x3*y3 + 4*x4*y4, GRB.MAXIMIZE)

# Add constraints
# m.addConstr(x1 <= M*y1, "c1")  # constraint to set x1 in optimization function 
m.addConstr(600 - x1 <= M*(1 - y1), "c2") # constraint for the 600 set (would prefer x >= 600 + M)
# m.addConstr(x2 <= M*y2, "c3")
m.addConstr(600 - x2 <= M*(1 - y2), "c4")
# m.addConstr(x3 <= M*y3, "c5")
m.addConstr(600 - x3 <= M*(1 - y3), "c6")
# m.addConstr(x4 <= M*y4, "c7")
m.addConstr(600 - x4 <= M*(1 - y4), "c8")
m.addConstr(90*x1 + 125*x2 + 175*x3 + 250*x4 <= 65000, "DoughConstraint")
m.addConstr(35*x1 + 50*x2 + 75*x3 + 90*x4 <= 25000, "IcingConstraint")
m.addConstr(x1 + x2 + x3 + x4 <= 800, "LaborConstraint")
# Solve model
m.optimize()
# Print solution
if m.status == GRB.Status.OPTIMAL:
    print('Optimal␣solution␣found:')
    for v in m.getVars():
        print(f'{v.varName}:␣{v.x}')
    else:
        print('No␣optimal␣solution␣found')
```

### Question 5

```python
m = gp.Model('question5')

h = m.addVar(vtype=gp.GRB.INTEGER, name='houses')
a = m.addVar(vtype=gp.GRB.INTEGER, name='apartments')
y1 = m.addVar(vtype=gp.GRB.BINARY, name='Choose Swimming-Tennis')
y2 = m.addVar(vtype=gp.GRB.BINARY, name='Choose Marina')

m.setObjective(6*h + 8*a - 1200*y1 - 2800*y2, gp.GRB.MAXIMIZE)

M = 1e7

# need to select either the marina or the swimming-tennis
m.addConstr(y1 + y2 == 1)

# totla site can accomdate 10k units
m.addConstr(h + a <= 10000)

# conditional logic for the house:apartment ratio

# # BIG M-Method (drag allow for x-to persist normally)
# m.addConstr(h >= 3 * a - M *(1-y1))

# non-BIG M-Method (yi works in {0,1} since the lower limit is 1 and upper is the boundary 3a)
m.addConstr(h >= 3 * a *y1)

# m.addConstr(h >= 0)
# m.addConstr(a >= 0)

m.optimize()

if m.status == gp.GRB.OPTIMAL:
    
    print(f'houses: {h.x}')
    print(f'apartment: {a.x}')
    print(f'Swimming-Tennis: {y1.x}')
    print(f'Marina: {y2.x}')
    print(m.ObjVal)
```

### Question 6

Function $I$ is Indian Journal and $F$ Fenghua Shibao number of exposures for a given number $x$ and $y$ for each respectively

$$I(x)+ F(y)$$

$$I(8)= 6(10) + 2(3) = 66$$


```python
def I(x):
    
    if x <= 6:
        return x*10
    elif x <= 10:
        return 6*10 + (x-6)*3
    elif x <= 15:
        return 6*10 + 4*3 + (x-10)*2.5
    else:
        return 6*10 + 4*3 + 5*2.5
    
```

```python
def F(x):
    
    if x <= 4:
        return x*8
    elif x <= 12:
        return 4*8 + (x-4)*6
    elif x <= 15:
        return 4*8 + 8*6 + (x-12)*2
    else:
        return 4*8 + 8*6 + 3*2.5
```

```python
m = gp.Model('question6')

x1 = m.addVar(vtype=gp.GRB.INTEGER, name='ij_1')
x2 = m.addVar(vtype=gp.GRB.INTEGER, name='ij_2')
x3 = m.addVar(vtype=gp.GRB.INTEGER, name='ij_3')

y1 = m.addVar(vtype=gp.GRB.INTEGER, name='fs_1')
y2 = m.addVar(vtype=gp.GRB.INTEGER, name='fs_2')
y3 = m.addVar(vtype=gp.GRB.INTEGER, name='fs_3')

# when a sub-group problem exists with more than 2 conditions, we break into sub-group 
m.setObjective(10*x1+3*x2+2.5*x3 + 8*y1+6*y2+2*y3, gp.GRB.MAXIMIZE)

# need to select either the marina or the swimming-tennis
m.addConstr(x1 <= 6)
m.addConstr(x2 <= 4)
m.addConstr(x3 <= 5)
m.addConstr(y1 <= 4)
m.addConstr(y2 <= 8)
m.addConstr(y3 <= 3)

m.addConstr(x1 + x2 + x3 + y1 + y2 + y3 <= 20)

m.optimize()

if m.status == gp.GRB.OPTIMAL:
    
    print(f'{x1.x}, {x2.x}, {x3.x}')
    print(f'{y1.x}, {y2.x}, {y3.x}')
```

## Question 7

```python
benefit = {1:65, 2:80, 3:30}
weight = {1:2, 2:3, 3:1}
```

```python
m = gp.Model('question6')

alloc = {i: m.addVar(vtype=gp.GRB.INTEGER, name=f'item_{i}') for i in range(1, 3+1)}

# when a sub-group problem exists with more than 2 conditions, we break into sub-group 
m.setObjective(sum(benefit[i]*alloc[i] for i in range(1, 3+1)), gp.GRB.MAXIMIZE)

m.addConstr(sum(alloc[i]*weight[i] for i in alloc) <= 5)

m.optimize()

if m.status == gp.GRB.OPTIMAL:
    
    for i in alloc:
        print(f'{alloc[i].x}')
```

### TSP Problem 

```python
np.random.seed(33)

#These are x and y coordinates, stored as different lists, for 15 points (they could be thought of as latitude / longitudes)
# x_coordinates = [1, 3, 5, 7, 2, 8, 7, 3, 1.5, 4, 6, 3, 4, 6, 5]
# y_coordinates = [2, 1, 3, 2, 4, 3, 4, 5, 3, 2, 2.5, 3, 4, 3.5, 5]

x_coordinates = [1, 3, 5, 7]
y_coordinates = [2, 1, 3, 2]

Distance=[]

#This computes the Euclidean distance between all pairs of points
Distance=[[round(((x_coordinates[i]-x_coordinates[j])**2+(y_coordinates[i]-y_coordinates[j])**2)**0.5,2) 
           for j in range(len(x_coordinates))] for i in range(len(x_coordinates))]

n = len(Distance)
```

```python
Distance
```

```python
m = Model("TSP")

# define the first set of binary allocation variables
alloc = {f'{i},{j}': m.addVar(vtype=gp.GRB.BINARY, name=f'({i},{j})') for i in range(n) for j in range(n)}

# set the objective function to be the binary x the distance value from nodes
m.setObjective(sum(alloc[p] * Distance[int(p.split(',')[0])][int(p.split(',')[1])] for p in alloc), gp.GRB.MINIMIZE)

u = {}
for i in range(n):
    # because index at 0 we start with 0+1 or 1 for lower bound and n-1 for n constraint in MTZ
    # if starting is different that first node 0 we initialize the point list and continue as normal
    if (i > 0):
        u[i] = m.addVar(vtype=GRB.CONTINUOUS, lb=1.0, ub=n-1, obj=0.0, name=f'u({i})' ) 
    else:
        u[i] = m.addVar(vtype=GRB.CONTINUOUS, lb=0, ub=0, obj=0.0, name=f'u({i})' )
        #The above assigns u[0] to zero via its lb and ub

#Constraints
#1: Each job has an arriving arc (going in flow)
for j in range(n):
    m.addConstr(sum(alloc[f'{i},{j}'] for i in range(n)) == 1)

#2: Each job has a departing arc (going out flow)
for i in range(n):
    m.addConstr(sum(alloc[f'{i},{j}'] for j in range(n)) == 1)
#     alloc[f'{i},{i}'].ub = 0
    
# a node can't visit itself (no self reference) 
for i in range(n):
    m.addConstr(alloc[f'{i},{i}'] == 0) 

#3: Use MTZ constraints to forbid subtours
for i in range(n-1):
    for j in range(n-1):
        if (i != j):
            # if path is choosen (i.e. x=1) then our next step Uj is shown to be the succesive iterative in sequence
            # if path is not choosen (i.e. x=0) then our step can be any of the n-1 nodes available
            m.addConstr(u[j+1] >= u[i+1] + 1 - (n-1)*(1- alloc[f'{i+1},{j+1}']))

#solve
m.optimize()

# Print optimal sequence of TSP nodes (these are stored in the u variables)
print('\nThe optimal sequence is:')
u_order_vector = [int(u[i].x) for i in range(n)]
print('\tU Order vector ', u_order_vector)
points = list(zip(x_coordinates, y_coordinates))
tsp_points = [points[i] for i in u_order_vector]
print('\tData Points (nodes) ', tsp_points)
# print('\tNode Distance ', [points[i] for i in tsp_points])
```
