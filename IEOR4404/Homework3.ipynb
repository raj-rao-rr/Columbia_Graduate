### Problem 2

```python
def method1():
    """Inverse Transform Method"""
    return np.log(1 + np.random.uniform() * ( np.exp(1) - 1) )

def method2():
    """Acceptance-Rejection Method"""
    a1 = np.random.uniform()
    a2 = np.random.uniform()
    
    if a2 <= np.exp(a1-1):
        return a1
    else:
        return method2()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].hist(x=[method1() for i in range(100)], 
           bins=40, alpha=0.5)
ax[1].hist(x=[method2() for i in range(100)], 
           bins=40, alpha=0.5)
plt.show()
```

### Problem 3

```python
import math
```

```python
def func(n, x):
    return (2*x**(n-1)) / (np.exp(x/2) * math.factorial(n-1)) 
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

x = np.arange(0, 11)

for n in range(1, 10):
    y = [func(n=n, x=i) for i in x]
    
    ax.plot(x, y, label=f'n={n}')

plt.legend()
plt.show()
```
